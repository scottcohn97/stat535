```python
# init game params

numAgents = 100
mu = 10 
giniArr = [0.1, 0.3, 0.5, 0.7, 0.9]
numSims = 25
prob1 = [0.15, 0.10, 0.25, 0.50]
prob2 = [0.35, 0.10, 0.25, 0.30]
prob3 = [0.35, 0.05, 0.25, 0.35]
```

## High Cooperation 1

```python
# create game instance `g`

g1 = Game(numAgent = numAgents, Gini = giniArr[0], para = 2, prob = prob1) # Gini = 0.1
g3 = Game(numAgent = numAgents, Gini = giniArr[1], para = 2, prob = prob1) # Gini = 0.3

g7 = Game(numAgent = numAgents, Gini = giniArr[3], para = 2, prob = prob1) # Gini = 0.1
g9 = Game(numAgent = numAgents, Gini = giniArr[4], para = 2, prob = prob1) # Gini = 0.3
```

```python
# Check distro is correct
strat_distribution(g1, numAgents)
```

```python
# run sims for each game instance

sim(g1, numSims)
sim(g3, numSims)

sim(g7, numSims)
sim(g9, numSims)
```

```python
# Put games in df 

g1_df = game_df(g1, numAgents)
g3_df = game_df(g3, numAgents)
g7_df = game_df(g7, numAgents)
g9_df = game_df(g9, numAgents)
```

```python
# write df to csv

g1_df.to_csv('g1_hc1.csv', header=True, index=False) 
g3_df.to_csv('g3_hc1.csv', header=True, index=False) 
g7_df.to_csv('g7_hc1.csv', header=True, index=False) 
g9_df.to_csv('g9_hc1.csv', header=True, index=False) 
```
